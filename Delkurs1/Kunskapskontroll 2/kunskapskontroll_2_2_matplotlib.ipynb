### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

##### Explicit, object oriented and implicit which is a simpler method, it is recommended to use the explicit method

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

##### Figure is the main object that contains all the parts for drawing graphs etc
##### Axes is the drawing area in the figure where the data is plotted and contains axes, labels, etc
##### Axis is the x and y axes of the graph.
##### Artist are objects that are drawn in Matplotlib, such as lines, text, axes, etc.

### When plotting in Matplotlib, what is the expected input data type?

##### Plotting functions expect numpy.array or numpy.ma.masked_array as input

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

### Pyplot approach

```python
x = np.linspace(-4, 4, 10)  # Sample data.

plt.figure(figsize=(3, 3), layout='constrained')
plt.plot(x, x**2, label='quadratic')  # etc.

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Object-oriented approach

```python
x = np.linspace(-4, 4, 10)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(3, 3), layout='constrained')

ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...

ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```
